```python
import pandas as pd
import geopandas 
```

```python
path = geopandas.datasets.get_path('naturalearth_lowres')
df = geopandas.read_file(path)
```

```python
df.plot()
```

```python
df.head()
```

```python
df['gdp_per_head'] = df.gdp_md_est / df.pop_est
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
fig, ax = plt.subplots(1,1)
divider = make_axes_locatable(ax)
cax = divider.append_axes('right', size='5%',pad=0.1)
df.plot(column = 'gdp_per_head', ax=ax, legend = True, cax = cax)
```

```python

fig, ax = plt.subplots(1,1)
divider = make_axes_locatable(ax)

df.plot(
    column = 'gdp_per_head', 
    cmap='OrRd', # choosing color
        ax=ax, 
        legend = True, 
        legend_kwds={'label': "Population by Country",'orientation': "horizontal"}
       )
```

```python
# how to handle missing value
import numpy as np
df.loc[np.random.choice(df.index, 40), 'pop_est']=np.nan
df.plot(column = 'pop_est', 
    cmap='OrRd', # choosing color
    legend = True, 
    figsize = (15,10),
        legend_kwds={'label': "Population by Country",'orientation': "horizontal"},
    missing_kwds={ "color": "lightgrey", "edgecolor": "red","hatch": "///","label": "Missing values",}
       )
```
