标准正交基有很多好处，如何能将非标准正交基变换为标准正交基呢？施密特正交化

施密特正交化就是把非正交基：
$\boldsymbol{x_1},\boldsymbol{x_2},\boldsymbol{x_3},\cdots,\boldsymbol{x_n}$

$通过正交化，得到正交基（下面用\boldsymbol{v_i}来表示正交基）：$

$\boldsymbol{v_1},\boldsymbol{v_2},\boldsymbol{v_3},\cdots,\boldsymbol{v_n}$

但是容易举一反三正交化的过程：

$\text{任选n-1个向量正交化}\to\text{第n个向量往这n-1个向量形成的体作垂线}$

从而完成正交化，下面来看看实现过程

有两个线性无关的向量：

$\boldsymbol{x_1},\boldsymbol{x_2}$

正交化的目标：

$\boldsymbol{x_1},\boldsymbol{x_2}\xrightarrow{\quad\text{施密特正交化}\quad}\boldsymbol{v_1},\boldsymbol{v_2}$

图示如下：

<img src = "image/ev15.png">

根据向量计算法则，可以知道：

$\boldsymbol{v_2}=\boldsymbol{x_2}-(\boldsymbol{x_2}\text{在}\boldsymbol{v_1}\text{上的投影向量})$

投影向量可以如下计算（根据点积的意义容易得到）：

$\boldsymbol{x_2}\text{在}\boldsymbol{v_1}\text{上的投影向量}=\frac{\boldsymbol{x_2}\cdot\boldsymbol{v_1}}{\boldsymbol{v_1}\cdot\boldsymbol{v_1}}\boldsymbol{v_1}$

所以：


$\begin{aligned}
    \boldsymbol{v_2}
        &=\boldsymbol{x_2}-(\boldsymbol{x_2}\text{在}\boldsymbol{v_1}\text{上的投影向量})\\
        \quad\\
        &=\boldsymbol{x_2}-\frac{\boldsymbol{x_2}\cdot\boldsymbol{v_1}}{\boldsymbol{v_1}\cdot\boldsymbol{v_1}}\boldsymbol{v_1}
\end{aligned}$



至此，完成任务：

$\boldsymbol{x_1},\boldsymbol{x_2}
\xrightarrow{\quad\text{施密特正交化}\quad}
\begin{cases}
    \boldsymbol{v_1}=\boldsymbol{x_1}\\
    \quad\\
    \boldsymbol{v_2}=\boldsymbol{x_2}-\frac{\boldsymbol{x_2}\cdot\boldsymbol{v_1}}{\boldsymbol{v_1}\cdot\boldsymbol{v_1}}\boldsymbol{v_1}
\end{cases}$

得到了正交基：

<img src = "image/ev19.png">


有三个线性无关的向量：

\boldsymbol{x_1},\boldsymbol{x_2},\boldsymbol{x_3}
$\boldsymbol{x_1},\boldsymbol{x_2},\boldsymbol{x_3}
\xrightarrow{\quad\text{施密特正交化}\quad}
\begin{cases}
    \boldsymbol{v_1}=\boldsymbol{x_1}\\
    \quad\\
    \boldsymbol{v_2}=\boldsymbol{x_2}-\frac{\boldsymbol{x_2}\cdot\boldsymbol{v_1}}{\boldsymbol{v_1}\cdot\boldsymbol{v_1}}\boldsymbol{v_1}\\
    \quad\\
    \boldsymbol{v_3}=\boldsymbol{x_3}-\frac{\boldsymbol{x_3}\cdot\boldsymbol{v_1}}{\boldsymbol{v_1}\cdot\boldsymbol{v_1}}\boldsymbol{v_1}-\frac{\boldsymbol{x_3}\cdot\boldsymbol{v_2}}{\boldsymbol{v_2}\cdot\boldsymbol{v_2}}\boldsymbol{v_2}
\end{cases}$


举一反三：

$\boldsymbol{x_1},\boldsymbol{x_2},\boldsymbol{x_3}\cdots,\boldsymbol{x_n}
\xrightarrow{\quad\text{施密特正交化}\quad}
\begin{cases}
    \boldsymbol{v_1}=\boldsymbol{x_1}\\
    \quad\\
    \boldsymbol{v_2}=\boldsymbol{x_2}-\frac{\boldsymbol{x_2}\cdot\boldsymbol{v_1}}{\boldsymbol{v_1}\cdot\boldsymbol{v_1}}\boldsymbol{v_1}\\
    \quad\\
    \boldsymbol{v_3}=\boldsymbol{x_3}-\frac{\boldsymbol{x_3}\cdot\boldsymbol{v_1}}{\boldsymbol{v_1}\cdot\boldsymbol{v_1}}\boldsymbol{v_1}-\frac{\boldsymbol{x_3}\cdot\boldsymbol{v_2}}{\boldsymbol{v_2}\cdot\boldsymbol{v_2}}\boldsymbol{v_2}\\
    \quad\\
    \boldsymbol{v_n}=\boldsymbol{x_n}-\frac{\boldsymbol{x_n}\cdot\boldsymbol{v_1}}{\boldsymbol{v_1}\cdot\boldsymbol{v_1}}\boldsymbol{v_1}-\frac{\boldsymbol{x_n}\cdot\boldsymbol{v_2}}{\boldsymbol{v_2}\cdot\boldsymbol{v_2}}\boldsymbol{v_2}-\cdots-\frac{\boldsymbol{x_n}\cdot\boldsymbol{v_{n-1}}}{\boldsymbol{v_{n-1}}\cdot\boldsymbol{v_{n-1}}}\boldsymbol{v_{n-1}}
\end{cases}$

然后，把每个向量进行单位化就可以得到标准正交基：


$\boldsymbol{v_1},\boldsymbol{v_2},\boldsymbol{v_3},\cdots,\boldsymbol{v_n}\xrightarrow{\quad\text{单位化}\quad}\boldsymbol{\epsilon_1},\boldsymbol{\epsilon_2},\boldsymbol{\epsilon_3},\cdots,\boldsymbol{\epsilon_n}$


前面我们介绍了两种办法获得较“好”的基：

对角化：以特征向量为基
正交化：以单位正交向量为基

既对角又正交的矩阵称对角正交矩阵。有没有矩阵一定可以对角正交化呢？有的，对称矩阵就可以。

如果矩阵A是对称矩阵，即：

$A=A^\mathrm{T}$

<img src = "image/ev20.png">

$上图中，\Lambda为对角阵\boldsymbol{p_1},\boldsymbol{p_2}为特征向量。$

$A=A^\mathrm{T}\iff A可对角正交化$

如果A是对称矩阵，那么不同特征向量空间的任意两个特征向量是正交的。
