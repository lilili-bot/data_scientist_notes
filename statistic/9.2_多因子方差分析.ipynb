1.双因子方差分析
多因子方差分析：用来判断由两个或两个以上名义变量分割的多个总体的均值是不是相等,可用于分析不同品牌和不同地区对一种商品销售量是否影响等

有两个因子时叫双因子方差分析

1.1 三组假设

第一个因子：

- $H_{01}: 第一个因素各⽔平的效应都为0$
- $H_{11}: 第一个因素各⽔平的效应不全为0$

第二个因子：

- $H_{02}: 第二个因素各⽔平的效应都为0$
- $H_{12}: 第二个因素各⽔平的效应不全为0$

两个因子的交互作用：

- $H_{03}: 二个因素交互效应都为0$
- $H_{13}: 二个因素交互效应不全为0$
  - 多因子方差分析可分为无交互作用和有交互作用两种。

  - 一个因子水平下的指标好坏及程度不受另一个因子不同水平的影响，则称两因子之间无交互作用，要分别判断每个因子对指标的影响。
  - 一个因子水平下的指标好坏及程度与另一个因子取什么水平有关，则称两因子之间有交互作用，记作A×B。此时除了要分别考虑每个因子对指标的影响外，还要考虑两个因子的搭配对指标产生新的效应。
  
  
1.2 双因素⽅方差分析的前提条件

- 独⽴立性:
  - (1) 组内独立(随机抽样/分配; 样本容量 < 10%总体容量);
  - (2) 组间独立(⾮配对)
- 正态性: 各组总体服从正态分布
  - 样本容量较大(每组样本容量 ≥ 10)时, 如果⼀定程度上违反了正态性, 仍可以使用ANOVA
  - 样本容量较小时, 如果违反了了正态性, 则应使用非参数方法进行分析
- ⽅差⻬性: 各组总体的方差相等
  - 各组样本的样本容量相等时, 如果⼀定程度上违反了⽅差齐性， 仍可以使用ANOVA
  - 各组样本的样本容量不相等时, 如果最大的样本标准差与最小的样本标准差之⽐不超过2, 仍可以使用ANOVA

```python
#无交互作用方差分析
import pandas as pd
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
```

```python
df=pd.read_csv("../data/f_twoway.csv")
```

```python
#最小二乘法拟合
model = ols('Score ~ Method+Reward',df).fit()
#到处方差分析结果
anovat = anova_lm(model)
anovat
```

```python
#有交互作用方差分析
#最小二乘法拟合
model = ols('Score ~ Method+Reward+Method*Reward',df).fit()
#到处方差分析结果
anovat = anova_lm(model)
anovat
```

- 多因子方差分析：用来判断由两个或两个以上名义变量分割的多个总体的均值是不是相等。
- 效应：
  - 主效应：因子本身对因变量均值的影响
  - 交互效应：因子交互作用(剔除主效应)对因变量均值的影响
- $MS_R,MS_C,MS_{R\times C}$
