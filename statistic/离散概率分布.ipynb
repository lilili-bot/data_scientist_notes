```python
import scipy.stats as ss
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def plotting(X,p_list, title):
    plt.plot(X,p_list, marker="o")
    plt.xlabel("x")
    plt.ylabel("p(x)")
    plt.title("%s"%title)
    plt.show()
```

# 离散概率分布
 ## 1.1 伯努利分布

```python
#定义随机变量。伯努利的随机变量取值只能是0，1。
X= np.arange(0,2) 
```

```python
# 设置概率
p = 0.5
#返回概率分布
p_list = ss.bernoulli.pmf(X,p)
```

```python
title = "bernoulli"
plotting(X,p_list, title=title)
```

## 1.2 二项分布

$X\sim b(n,p)$

$p(k)=P(X=k)={n\choose k}p^k(1-p)^{n-k},\quad k=0,1,\cdots,n$

期望=np （预期成功多少次）

方差=np(1-p)（数据波动大小）


```python
# 随机事件：抛硬币五次，正面朝上的次数。
#定义随机变量, 
n=5
p = 0.5
# 生成随机变量的样本空间
X = np.arange(0,6,1)
```

```python
#输出概率质量（函数）分布
p_list= ss.binom.pmf(X,n,p) 
```

```python
title = "binom with n is %d and p is %d"%(n,p)
plotting(X,p_list, title=title)
```

## 1.3.1 泊松分布

1）事件是独立事件 （跟赌徒谬论一样，每次抽奖都是独立事件）

2）在任意相同的时间范围内，事件发生的概率相同 （例如1天内中奖概率，与第2天内中间概率相同） 

3）你想知道某个时间范围内，发生某件事情x次的概率是多大 （例如你搞了个促销抽奖活动，想知道一天内10人中奖的概率）

对于随机变量X的概率质量函数：

$P(X=k)=\frac{\lambda^k}{k!}e^{-\lambda},\quad k=0,1,2,\cdots$

$称为随机变量X的\color{Salmon}{泊松分布}，也可以记为：$

$X\sim P(\lambda)$

其数学期望和方差为：

$E(X)=\lambda,\quad Var(X)=\lambda$

## 例：
已知一天内某路口平均每天发生2次事故。求该路口一天内发生4次事故的概率。

```python
# 1. 定义分布参数：均值 mu ，需求的‘随机事件发生的’次数 k
mu = 2
# 定义随机变量的取值。注意泊松分布中，随机变量的取值可以是无穷大的。
k = 100
X = np.arange(0,k+1)
```

```python
# 输出概率
p_list = ss.poisson.pmf(X,mu)
```

```python
title= "poisson PMF when the mu is %d"%mu
plotting(X,p_list, title=title)
```

## 1.3.2 泊松分布扩展
泊松分布实在限制太大，只告诉了我们每天卖出的馒头数。不过没有关系，稍微扩展下可以得到新的函数：

$P(X=k,t)=\frac{\left(\lambda t\right)^k}{k!}e^{-\lambda t}$

通过新的这个函数就可知不同的时间段内卖出的馒头数的分布了（t=1时就是泊松分布)

## 1.3.3 几何分布
两次卖出馒头之间的时间间隔大于t的概率，根据之前的分析，等同于t时间内没有卖出一个馒头的概率，而后者的概率可以由泊松过程给出。至此所需的条件都齐备了，那么开始解题吧，假设随机变量：

Y=两次卖出馒头之间的时间间隔

这个随机变量的概率可以如下计算：

$P(Y > t)=P(X=0, t)=\frac{\left(\lambda t\right)^0}{0!}e^{-\lambda t}=e^{-\lambda t},\quad t \ge 0$

进而有：

$P(Y \le t)=1-P(Y > t)=1-e^{-\lambda t}$

这其实已经得到了Y的累积分布函数了，所以我们可以得到一个概率分布：

若随机变量X的概率密度函数为：


$p(x)=\begin{cases}
\lambda e^{-\lambda x}, & x \ge 0\\
0,& x < 0
\end{cases}$

$其中\lambda > 0，称X服从\color{Salmon}{指数分布}，也可以记为：$

$X\sim Exp(\lambda)$

累积分布函数为：


$F(x)=\begin{cases}
1-e^{-\lambda x}, & x \ge 0\\
0,& x < 0
\end{cases}$


## 1.4 几何分布

对于n重伯努利实验，如果每次得到“是”的概率为p，设随机变量：
X=首次得到“是”时进行的试验次数

则称：

$p(k)=P(X=k)=(1-p)^{k-1}p,\quad k=1,2,\cdots$

$为随机变量X的\color{Salmon}{几何分布}，也可以记作:$

$X\sim Ge(p)$

其数学期望和方差为：

$E(X)=\frac{1}{p},\quad Var(X)=\frac{1-p}{p^2}$

### 首次表白成功的概率

```python
# 1. 定义分布参数。几何分布只需要每次成功的概率，不需要n
p = 0.5
# 2. 定义随机变量的取值。
X = np.arange(1,10)
```

```python
# 输出概率
p_list= ss.geom.pmf(X,p)
```

```python
title= "geom PMF when the p is %d"%p
plotting(X,p_list, title=title)
```
