```python
import numpy as np
import seaborn as sns
```

```python
# generate data
all_data = np.random.uniform(2,3,100000)
print("ZongTi mean is", all_data.mean())
print("ZongTi sd is", all_data.std())
```

```python
# plotting ZongTi steuerung.

def plothist(x, bins = 1000):
    sns.histplot(x, bins=bins)
```

```python
plothist(all_data, 1000)
```

```python
# function for sampling
def sampling(data, sample_size = 30, sample_amount = 10000):
    return np.array([np.random.choice(data, sample_size) for i in range(sample_amount)])
```

```python
# sample_size = 2
data2 = sampling(all_data,2)
data2.shape
```

```python
# calculate mean of each sample.
mu2 = data2.mean(axis = 1)
```

```python
# ploting mu2
plothist(mu2,1000)
```

```python
# sample_size = 5
data5 = sampling(all_data,5)
data5.shape
```

```python
# calculate mean of each sample.
mu5 = data5.mean(axis = 1)
```

```python
# ploting mu5
plothist(mu5,1000)
```

```python
# sample_size = 30
data30 = sampling(all_data,30)
data30.shape
# calculate mean of each sample.
mu30 = data30.mean(axis = 1)
# ploting mu30
plothist(mu30,1000)
```
