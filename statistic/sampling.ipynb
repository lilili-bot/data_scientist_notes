```python
import numpy as np
import seaborn as sns
```

```python
# generate data
all_data = np.random.uniform(2,3,100000)
print("ZongTi mean is", all_data.mean())
print("ZongTi sd is", all_data.std())
```

```python
# plotting ZongTi steuerung.

def plothist(x, bins = 1000):
    sns.histplot(x, bins=bins)
```

```python
plothist(all_data, 1000)
```

```python
# function for sampling
def sampling(data, sample_size = 30, sample_amount = 10000):
    return np.array([np.random.choice(data, sample_size) for i in range(sample_amount)])
```

```python
# sample_size = 2
data2 = sampling(all_data,2)
data2.shape
```

```python
# calculate mean of each sample.
mu2 = data2.mean(axis = 1)
```

```python
# ploting mu2
plothist(mu2,1000)
```

```python
# sample_size = 5
data5 = sampling(all_data,5)
data5.shape
```

```python
# calculate mean of each sample.
mu5 = data5.mean(axis = 1)
```

```python
# ploting mu5
plothist(mu5,1000)
```

```python
# sample_size = 30
data30 = sampling(all_data,30)
data30.shape
# calculate mean of each sample.
mu30 = data30.mean(axis = 1)
# ploting mu30
plothist(mu30,1000)
```

```python
# Conclusion: 随着样本容量不断增加，样本的均值的分布，越来越接近正态分布，并且抽样均值的分布分布在总体均值的周围。样本容量越大，抽样的样本的均值，越接近总体均值。  
# 30样本容量的标准误
mu30.std()
```

```python
# 总体的标准差
all_data.std()
```

```python
# 根据标准误的公式，用总体标准差/样本容量开方
se = all_data.std()/np.sqrt(30)
```

```python
se
# 结论，标准误的公式是合理的。标准误的计算有个两种方式： 1. 直接用抽样的每个样本的均值，求出标误差。2.总体的标准差/样本容量开根号。
```
