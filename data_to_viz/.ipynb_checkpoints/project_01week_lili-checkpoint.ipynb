```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

- step 1 Read files

```python
'''read file'''
fert = pd.read_csv('data/gapminder_total_fertility.csv')
```

```python
'''
read excel. There was a bug, unsing read_excel will not work. 
Solution is from Stackoverflow
https://stackoverflow.com/questions/65254535/xlrd-biffh-xlrderror-excel-xlsx-file-not-supported
'''
life = pd.read_excel(
     os.path.join("data", "gapminder_lifeexpectancy.xlsx"),
     engine='openpyxl',
)
```

```python
life.info()
```

```python
'''pre-work on data'''
fert=fert.rename(columns={'Total fertility rate':'1'})
```

```python
fert.columns=[int(i) for i in fert.columns]
```

```python
fert=fert.rename(columns={1:'country'})
```

```python
fert=fert.melt(id_vars='country', var_name='year', value_name='fertility_rate')
```

```python
life=life.rename(columns={'Life expectancy':'country'})
```

```python
life=life.melt(id_vars='country', var_name='year', value_name='lifeexp')
```

```python
life.head()
```

```python
fert.head()
```

```python
population= pd.read_excel(
     os.path.join("data", "gapminder_population.xlsx"),
     engine='openpyxl',
)
```

```python
population=population.rename(columns={'Total population':'country'})
```

```python
population=population.melt(id_vars='country', var_name='year', value_name='population')
```

```python
population.head()
```

```python
continents = pd.read_csv('data/continents.csv',decimal=';', sep = '\t')
```

```python
continents[['continent','country']]=continents['continent;country'].str.split(';', expand=True)
```

```python
continents.drop('continent;country', axis=1, inplace=True)
```

```python
continents.set_index('country',drop=True, inplace=True)
```

- step 2 merge data together and add given continent column

```python
df=fert.merge(population).merge(life)
```

```python
df['cont'] = None
for country in continents.index:
    df.loc[df['country']==country,'cont'] = continents.loc[country,'continent']
```

- def plot_frame(ax, df, year, limits=(1, 8.6, 27, 85.2)):
    """ limits: (xmin, xmax, ymin, ymax) """
    
    year = int(year)
    df_subset = df.loc[ df['year'].eq(year) ]
    
    sns.scatterplot(  x     = 'fertility_rate'
                    , y     = 'life_expectancy'
                    , hue   = 'continent'
                    , size  = 'population'
                    , sizes = (15,700)
                    , size_norm = (50,1_000_000_000)
                    , data  = df_subset
                    , alpha = 0.6
                    , ax    = ax)
    sns.despine(ax=ax)
    ax.legend(bbox_to_anchor=(1.1, 1),borderaxespad=0.1)
    ax.set_xlabel('Fertility Rate')
    ax.set_ylabel('Life Expectancy')
    ax.set_title('Year: {}'.format(year))
    
    if limits is not None:
        ax.axis(limits)
    return ax.axis() # return current limits

```python
for i in range(1960, 2016):
    df_to_gif = df.loc[df['year'] == i]
    plt.ylim(0, 87)
    plt.xlim(0, 10)
    plt.title(str(i))
    sns.scatterplot(data = df_to_gif, y= 'lifeexp', x= 'fertility_rate',
                    hue='country',
                    size = 'population', 
                    palette="Set2",
                    sizes =(60, 900),legend=False)
    a = df_to_gif.loc[df['country']=='China']['fertility_rate']
    b = df_to_gif.loc[df['country']=='China']['lifeexp']
    plt.annotate('China',color='orange', xy=(a, b),arrowprops=dict(facecolor='k', headwidth=5, width=0.5))
    c = df_to_gif.loc[df['country']=='Germany']['fertility_rate']
    d = df_to_gif.loc[df['country']=='Germany']['lifeexp']
    plt.annotate('Germany',color='blue', xy=(c, d), xytext=(c, d),arrowprops=dict(facecolor='k', headwidth=5, width=0.5))
    plt.savefig('lifeexp_%d.png'%(i))
    plt.close()
```

```python
'''plot year from 1960 to 2016 and mark Germany and China'''
country_list=['Germany','China']
start = 1960

'''create folder to save all images'''

path = 'png'
try: 
    os.makedirs(path)
except OSError:
    print ("Creation of the directory %s failed" % path)
else:
    print ("Successfully created the directory %s" % path)
    
'''start plotting'''   

while start < 2016:
    df_to_plot = df.loc[df['year'] == start]
    fig = plt.figure(figsize = (20,10))
    ax = fig.add_subplot()
    ax.set_facecolor('#2c2c2c')
    ax.grid(False)
    ax.annotate(text = str(start), xy= (13,9), fontsize =30, color = 'white')
    for i in country_list:
        x_cor = float(df_to_plot.loc[df_to_plot['country']== i, ['lifeexp']].values[0])
        y_cor = float(df_to_plot.loc[df_to_plot['country']== i, ['fertility_rate']].values[0])
        ax.annotate(text = i, xy= (x_cor,y_cor), fontsize =15, color = 'white', weight = 'bold', arrowprops={})
    plt.axis((10, 90, 0, 10))
    sns.axes_style('white')
    fig.set_facecolor('#2c2c2c')
    sns.despine()
    plt.xticks(fontsize=10, color = 'white')
    plt.xlabel('Life Expectancy', color ='white')
    plt.ylabel('Fertility', color = 'white')
    plt.yticks(fontsize=10, color = 'white')
    sns.scatterplot( data = df_to_plot, x= 'lifeexp', y = 'fertility_rate', hue='cont', size ='population', sizes = (50,600),palette= "Set2")
    plt.savefig('png/lifeexp_'+str(start)+'.png')
    plt.close()
    start = start + 1
```

```python
'''create gif with imageio'''
import imageio

images = []

for i in range(1960, 2016):
    filename = 'png/lifeexp_{}.png'.format(i)
    images.append(imageio.imread(filename))

imageio.mimsave('output_1.gif', images, fps=10)
```
