given vector a and b, what is the shortest distance from a to b?

```python
'''visulize the question'''
fig = plt.figure()
'''set 3d ax'''
ax = fig.gca(projection = '3d')
'start and end point of the vector'
a = ax.quiver(0,0,0,1,2,3, arrow_length_ratio=0.3, normalize = False)
b = ax.quiver(0,0,0,1,1,1, arrow_length_ratio=0.3, normalize = False)
'config x,y,z'
ax.set_xlim([0,5])
ax.set_ylim([0,5])
ax.set_zlim([0,5])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
'text on vector'
ax.text(1,2,3,'a',color='red')
ax.text(1,1,1,'b',color='blue')    
```

analysis the question: the showrtest distance from a to b, is e
    $$ e = \vec{a} - Proj_\vec{b} \vec{a} $$
    $$ \vec {p} = \hat{x}  \vec{b} = Proj_\vec{b} \vec{a}$$
    $$ \vec {b} e = \vec {b}  (\vec{a}-\hat{x}  \vec{b}) = 0$$
    $$ \hat{x} = \frac{\vec {b}*\vec{a}}{\vec{b}*\vec{b}}$$
    $$ \hat{x} \vec{b} = \vec{b}*\hat{x} = \vec{b}\frac{\vec {a}*\vec{b}}{\vec{b}*\vec{b}} 
    = \vec{b}\frac{\vec{b}^T \vec {a}}{\vec{b}^T*\vec{b}} 
    = \frac{\vec{b}^T \vec{b}}{\vec{b}^T\vec{b}}\vec {a} $$
    $$  P = \frac{\vec{b}^T \vec{b}}{\vec{b}\vec{b}^T} $$
the question is now to find,$$ p,  \hat{x} , MatrixP $$

```python
a_vec = np.array([1,2,3])
```

```python
b_vec = np.array([1,1,1])
```

```python
'''calculate x_hat'''
x_hat = np.dot(a_vec.T,b_vec)/np.dot(b_vec.T,b_vec)
x_hat
```

```python
'''calculate p based on x_hat value'''
p = np.dot(x_hat,b_vec)
p
```

```python
'''calculate projection matrix'''
P = a_vec.T * a_vec/ a_vec *a_vec.T
```
